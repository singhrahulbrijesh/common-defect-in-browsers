```python
import re
def update_component(row):
    # Check if the component column is a string
    if isinstance(row['component'], str):
        components = re.split(r'[>,]', row['component'])
        components = [component.lower().strip() for component in components]
        # Iterate over the components in reverse order
        for component in reversed(components):
            # Look for a matching name in df2
            match = df2[df2['name'].str.lower() == component]
            if not match.empty:
                # If a match is found, update the component_new column with the matched name
                row['component_new'] = match.iloc[0]['name']
                break  # Stop iterating once a match is found
    return row
ch_bug_data = ch_bug_data.apply(update_component, axis=1)

ch_bug_data.head(30)
```

```python
# Create a dictionary mapping df2 names to comp_id
name_to_comp_id = df2.set_index('name')['id'].to_dict()

# Map the comp_id based on component_new
ch_bug_data['comp_id'] = ch_bug_data['component_new'].map(name_to_comp_id)

# Print the updated ch_bug_data dataframe
print(ch_bug_data)
```
